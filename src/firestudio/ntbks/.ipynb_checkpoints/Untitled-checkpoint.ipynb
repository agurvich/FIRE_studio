```python
%load_ext autoreload
%autoreload 2
import numpy as np
from firestudio.studios.gas_studio import GasStudio
from abg_python.snapshot_utils import openSnapshot
from abg_python.cosmoExtractor import extractDiskFromSnapdict,offsetRotateSnapshot
from abg_python.cosmo_utils import load_AHF
from abg_python.all_utils import filterDictionary
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
snapdir = "/Users/agurvich/research/snaps/m12i_res7100/output/"
snapnum = 600 
gas_snap = openSnapshot(snapdir,snapnum,0)
star_snap = openSnapshot(snapdir,snapnum,4)
```

```python
scom,rvir,rstar_half = load_AHF(snapdir,snapnum,0)
```

```python
thetay,thetaz,scom,vscom,gindices,sindices,radius = extractDiskFromSnapdict(
    star_snap,gas_snap,25,scom=scom,)
```

```python
sub_gas_snap = offsetRotateSnapshot(gas_snap,scom,vscom,thetay,thetaz,orient_stars = 0)
xs,ys = gas_snap['Coordinates'][:,:2].T
gindices = np.logical_and(np.abs(xs)<=25,np.abs(ys)<=25)
sub_gas_snap = filterDictionary(
    gas_snap,
    gindices)
xs,ys = sub_gas_snap['Coordinates'][:,:2].T
```

```python
from matplotlib.colors import LogNorm
plt.hist2d(xs,ys,bins=1000,norm=LogNorm())
plt.gca().set_aspect(1)
```

```python
my_studio = GasStudio(
    snapdir,
    snapnum,
    '/Users/agurvich/research/repos/FIRE_studio/',
    frame_half_width=25,
    frame_depth=25,
    snapdict=sub_gas_snap,
    )
```

```python
my_studio.render(plt.gca(),[
    'Temperature','foo'])
```

```python
rhats = sub_gas_snap['Coordinates']/np.linalg.norm(sub_gas_snap['Coordinates'],axis=1)[:,None]
vrs = np.sum(rhats*sub_gas_snap['Velocities'],axis=1)
sub_gas_snap['Vr'] = vrs
plt.hist(vrs,bins=np.linspace(-100,100,50))
print()
```
